### loading the data and performing data validation and formating, this is the temperature dataset which was sccraped from the web and formating was done on it, like cleaning the data filling the missing data through interpolation method and then saving the result to csv named (temperature.csv)

#### The csv contain features like avg,max and min temperature, precipitation, snow etc. but as it is the gurgoan dataset there is no snowfall observed hence that whole column is full of Nan values.

```python
import pandas as pd
df=pd.read_csv(r'C:/Users/navee/Downloads/data.csv',skiprows=1)
df.Date=pd.to_datetime(df.Date)
df['Month']=df.Date.dt.month
df['Year']=df.Date.dt.year
```

### Save the dataset in the new csv file

```python
import numpy as np
m=df.Month==8
n=df.Month==9
o=df.Year==2019
x=df.loc[m | n]

m=x.Year==2019
df1=x.loc[m]

df1.drop(columns=['PRCP (Inches)','SNOW (Inches)','SNWD (Inches)'],inplace=True)
df1.rename(columns={'TAVG (Degrees Fahrenheit)':'avg_temp','TMAX (Degrees Fahrenheit)':'max_temp',
                    'TMIN (Degrees Fahrenheit)':'min_temp'},inplace=True)

df1.to_csv(r"C:\Users\navee\Desktop\temperature.csv")
```
