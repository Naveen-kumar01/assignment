### There are 35,000 hourly samples that are far too dense for us to make much sense of. We can gain more insight by resampling the data to a coarser grid. Let's resample by week:

```python
%matplotlib inline
import seaborn 
seaborn.set()
plt.rcParams['figure.figsize']=(20,10)
plt.style.use('ggplot')
df.plot()
plt.ylabel('Minutle-wise power Consumption')
```

```python
weekly = df.resample('W').sum()
weekly.plot()
plt.ylabel('power consumption')
```

### from above figure we can see that there is decline in the power consumption of the AC, in second week of the          month 1 and in first week of month 2

#### Other way that comes in handy for aggregating the dataset is to use a rolling mean, utilizing the pd.rolling_mean() function. Here           we'll do a 10 day rolling mean of our data, making sure to center the window

```python
daily = df.resample('D').sum()
daily.rolling(10, center=True).sum().plot(style=[':', '--', '-'])
plt.ylabel('mean hourly count')
```

#### The jaggedness of the result is due to the hard cutoff of the window. We can get a smoother version of a rolling mean using a                window function–for example, a Gaussian window. The following code specifies both the width of the window (we chose 10 days)          and the width of the Gaussian within the window (we chose 5 days)

```python
daily.rolling(10, center=True,win_type='gaussian').sum(std=5).plot(style=[':', '--', '-'])
```

### the hourly data also shows the variation of the data in every hours, but its too cluttered to understand anything so we need to scale up a bit.

```python
plt.rcParams['figure.figsize']=(20,10)
plt.style.use('ggplot')
by_time = df1.groupby(df1.index.time).sum()
hourly_ticks = 1 * 60 * 60 * np.arange(24)
by_time.plot(xticks=hourly_ticks, style=[':', '--', '-'])
```

#### The below graph shows the ac power consumption on weekdays and we can infer that the on thursday and on weekends there is an increase in the graph

```python
by_weekday = df1.groupby(df1.index.dayofweek).mean()
by_weekday.index = ['Mon', 'Tues', 'Wed', 'Thurs', 'Fri', 'Sat', 'Sun']
by_weekday.plot(style=[':', '--', '-']);
```

#### as we saw earlier that the consumption went mininum in the dates between 16 - 22, there can be many reason for such kind of anomaly behaviour
